```python
# Label encoding of target feature [class]
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
  

df1['class']= label_encoder.fit_transform(df1['class'])
  
df1['class'].unique()
```

```python
df1
```

```python
# defining case and control
case = df1[df1['class']==0]
control = df1[df1['class']==1]
```

```python
# finding all numeric columns

num_col = [col for col in df1.columns if df1[col].dtypes =='float']
```

```python
num_col
```

# Performing Student T-test using control and case

```python
from scipy.stats import ttest_ind
for col in num_col:
  stat, p = ttest_ind(case[col],control[col])
  print(f'P value for {col} : {p}')
  #print('stat= %.3f , p=%.3f ' % (stat,p))
```

```python
# Finding Log Fold Change
import math
# albumin and sugar not here coz min of this two columns is zero

li = ['age',
 'blood pressure',
 'specific gravity',
 'blood glucose random',
 'blood urea',
 'serum creatinine',
 'sodium',
 'potassium',
 'hemoglobin',
 'packed cell volume',
 'white blood cell count',
 'red blood cell count']
for col in li:
 lfc = math.log2(case[col].mean()) - math.log2(control[col].mean())
 print(f'LFC for {col}  : {lfc}')
```

```python
# Finding all categorical columns
cat_col = [col for col in df.columns if df[col].dtypes =='O']
cat_col
```

# Chi-Squared Test

```python
from scipy.stats import chi2_contingency
for col in cat_col:
  stat, p, dof, expected = chi2_contingency(pd.crosstab(df[col], df['class']))
  print(f'P value for {col} : {p}')
```
